## the graphs of odds, logit and sigmoid

```python
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
plt.style.use('seaborn')

p_np=np.linspace(0.01,0.99,100)
p_=np.linspace(0.01,0.99,100)

odds_np = p_np/(1-p_np)  # 일어나지 않을 확률에 비해서 일어날확률이 어느정도인지 -> 오즈
odds_tf = p_/(1-p_)

logit_np =np.log(odds_np) # log를 씌워서 0을 기준으로 대칭이 되도록 만든다 -> 로짓
logit_ =tf.math.log(odds_tf)

fig,axes=plt.subplots(2,1,figsize=(15,10), sharex=True)

axes[0].plot(p_np, odds_np)
axes[1].plot(p_np, logit_np)
xticks = np.arange(0,1.1,0.2)
axes[0].tick_params(labelsize=15)
axes[1].tick_params(labelsize=15)
```

## the graphs of sigmoid

```python
# 그리고 위 로짓을 p를 기준으로 역함수를 만들면 sigmoid가 된다.
# 모양은 아래와 같이 0~1사이의 확률값.
# 즉 sigmoid는 p(기냐), 1-p(아니냐)를 기준으로 만들어진 binary한 값이다
import tensorflow as tf
from tensorflow.keras.layers import Activation

X=tf.linspace(-10,10,100)
sigmoid=Activation('sigmoid')(X)

fix,ax = plt.subplots(figsize=(10,5))
ax.plot(X.numpy(), sigmoid.numpy())
```

## single-variate logistic regression model

```python
from tensorflow.keras.layers import Dense
plt.style.use('seaborn')

X= tf.random.normal(shape=(100,1))
dense=Dense(units=1, activation='sigmoid')
# sigmoid는 binary를 위한 p(확률값)으로 만드는것이기 때문에 units는 1)
Y=dense(X)
print(Y.shape)
fig,ax = plt.subplots(figsize=(7,7))
ax.scatter(X.numpy().flatten(), Y.numpy().flatten())
```

## multi-variate logistic regression model

```python
from tensorflow.keras.layers import Dense
plt.style.use('seaborn')

X= tf.random.normal(shape=(100,5))
dense=Dense(units=1, activation='sigmoid')
# sigmoid는 binary를 위한 p(확률값)으로 만드는것이기 때문에 units는 1)
Y=dense(X)
print(Y.shape)
```

## IO of softmax

```python
logit=tf.random.uniform(shape=(1,5), minval=-10, maxval=10)
softmax_value=Activation('softmax')(logit)
print(softmax_value)
softmax_sum=tf.reduce_sum(softmax_value,axis=1)
print(softmax_sum)
```

## softmax in Dense layers

```python
logit=tf.random.uniform(shape=(8,5), minval=-10, maxval=10)
dense=Dense(units=5, activation='softmax')
Y=dense(logit)
print(Y.shape)
tf.reduce_sum(Y,axis=1)
```

## multi-class classifier

```python
from tensorflow.keras.models import Model
```

```python
class TestModel(Model):
    def __init__(self):
        super(TestModel, self).__init__()
        
        self.dense1 = Dense(units=8, activation='relu')
        self.dense2 = Dense(units=5, activation='relu')
        self.dense3 = Dense(units=3, activation='softmax')
        
    def call(self, x):
        print('x',x.shape, X.numpy())
        
        x=self.dense1(x)
        print('a1',x.shape,x)
        x=self.dense2(x)
        print('a2',x.shape,x)
        x=self.dense3(x)
        print('y',x.shape,x)
        
        return x
```

```python
model=TestModel()

X=tf.random.uniform(shape=(8,5), minval=-10, maxval=10)
Y=model(X)
```

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, x):
        return self.factor * x

# Multiplier 클래스의 인스턴스 생성
double = Multiplier(2)

# 인스턴스를 함수처럼 호출
result = double(5)

print(result)  # 출력: 10
```

```python
import tensorflow as tf

class SimpleModel(tf.keras.Model):
    def __init__(self):
        super(SimpleModel, self).__init__()

    def call(self, inputs):
        return inputs * 2

# SimpleModel 클래스의 인스턴스 생성
model = SimpleModel()

# 모델을 호출하여 사용
result = model(5)

print(result)  # 출력: 10
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
function_x=np.linspace(-3,3,100)
function_x
```

```python
function_y=2*function_x**2

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(function_x,function_y)

x=3
lr=0.05

for _ in range(100):
    dy_dx= 4*x # 미분한값
    x=x-lr*dy_dx
    y=2*x**2
    
    ax.scatter(x,y, color='red', s=100)
    
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(function_x,function_y)

x=-2
lr=0.05

for _ in range(100):
    dy_dx= 4*x # 미분한값
    x=x-lr*dy_dx
    y=2*x**2
    
    ax.scatter(x,y, color='red', s=100)
```

```python
function_y=function_x**3

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(function_x,function_y)

x=3
lr=0.05

for _ in range(100):
    dy_dx= 3*x**2 # 미분한값
    x=x-lr*dy_dx
    y=x**3
    
    ax.scatter(x,y, color='red', s=100)
    
```

# real world
실제에서는 함수를 개형을 모르는 상태
따라서 미분계수만 가지고 gradient에 의존해서 학습하게됨 (gradient based learning)

```python
function_y=function_x**3

fig, ax = plt.subplots(figsize=(10,5))
# ax.plot(function_x,function_y)

x=3
lr=0.05

for _ in range(100):
    dy_dx= 3*x**2 # 미분한값 <- 이건어떻게알죠? 이거를 이제 손으로 계산하면서 보자 ㅎㅎ
    x=x-lr*dy_dx ### 결국 이렇게 업데이트 해주려고!!! 이거때문에 하는거임 이 모든게!!!
    y=x**3
    
    ax.scatter(x,y, color='red', s=100)
    
```

# chain rule

```python
from IPython.display import Image
image_path = '../images/chainrule.jpg'
Image(filename=image_path)
```

미분값은 분수가 아니다, 분수처럼 표현할뿐 <br>
분수처럼 사용할때가 있는데 그게 chain rule이다

합성함수에서 함수값을 구해나가는 방향과 chain rule을 적용하는(미분값을 구하는) 방향이 반대방향이라서 <br>
forward-backward, backpropagation이라고 지칭함

y변화에 대한 각 중간값이 미치는 영향도임 <br>
마지막 미분을 계산해야지 -1, -2 미분 계산가능하다 (함축하고 있기 때문에)

미분값을 구해야한다 <br>
loss값을 줄이기위해 어떤 방향으로 가야하는지 알기위해서

각 파라미터마다 전부 미부값이 필요하게되는거고 <br>
loss줄이는 방향으로 업데이트를 함 < 모든 w,b는 힘을합쳐서

이과정에서 chain rule을 활용해서 변수에 대한 미분을함

# why jacobians?

x vector 길이가 늘어나는만큼 W가 늘어난다 <br>
w1...wn에 대한 미분을 전부다 구해야함 <br>
백터에 대한 미분이 필요해짐

야코비안 행렬의 원소들은 모두 1차 미분 계수로 구성되어 있음 <br>
야코비안이 말하고자 하는 것은 ‘비선형 변환’을 ‘선형 변환으로 근사’시킨 것

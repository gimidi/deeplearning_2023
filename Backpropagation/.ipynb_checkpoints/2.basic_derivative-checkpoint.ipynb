# 로그함수의 미분

```python
대표적인 단순증가함수 (따라서 도함수가 항상 양수)
```

$$f(x)=\log_a x$$  <br>

$$f'(x)=\frac{1}{x  \times \log_e a}= \frac{1}{x} \frac{1}{\log_e a}$$ <br>

만약 a가 e라면.. $$f'(x)=\frac{1}{x}$$

# 지수함수의 미분

```python
power punction과 반대로 밑이 상수고 지수가 변수
로그함수와 역의관계 (로그의정의: a를 y제곱하면 x가 나온다, 따라서 x,y바꾸면 지수함수가 됨)
```

$$f(x)=a^x$$

$$f'(x)=a^x\log a$$

```python
따라서 자연상수를 밑으로하는 지수함수는 본인이 도함수가됨
원시함수와 도함수가 같은건 전체 수학에서 얘 뿐이다
```

$$f(x)=e^x$$

$$f'(x)=e^xloge=e^x$$

```python
특징이 원래함수에 상수배만큼 바뀌게 된다.
곱해지는 상수가 loga 이기 때문에
```

$$a^x\log a=f(x)\log a$$

# 삼각함수의 미분

```python
와 레전드네..
sin(x)의 도함수가 cos이 되어버리네..
유도과정..그냥 결과외울까해..
5번 이상 사용되면 공식해서 유도해보쟈..
```

$$f(x)=sin(x)$$

$$f'(x)=cos(x)$$

$$f(x)=cos(x)$$

$$f'(x)=-sin(x)$$

$$f(x)=tan(x)$$

$$f'(x)=\frac{1}{cos^2(x)}$$
